```python
#Import modules
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md

#Load Files (xtc)
traj = md.load_xtc('fit_unwrapped_water_nacl.xtc', top='step5_9.gro')

#Input function to calculate MSD for a given atom
def msd(traj, atom):
    # chlorine
    if atom == 0:
        return md.rmsd(traj, traj, 0)
    # sodium
    elif atom == 1:
        return md.rmsd(traj, traj, 1)
    # oxygen
    elif atom == 2:
        return md.rmsd(traj, traj, 2)

#Convert values to integers
msd_cl = msd(traj, 0).astype(int)
msd_na = msd(traj, 1).astype(int)
msd_o = msd(traj, 2).astype(int)

#Plot results vs time
plt.plot(msd_o, label='O')
plt.plot(msd_na, label='Na')
plt.plot(msd_cl, label='Cl')
plt.xlabel('Time (ps)')
plt.ylabel('MSD (nm)')
plt.legend()
plt.title('MSD of O, Na, & Cl vs Time')
plt.show()
```
