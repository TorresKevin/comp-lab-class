```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read in the data
nvt = np.loadtxt('../Outputs/NVT-1.ener', skiprows=1)
npt = np.loadtxt('../Outputs/NPT-1.ener', skiprows=1)

#plot the data
plt.plot(nvt[:,1], nvt[:,4], label='NVT')
plt.title('Potential Energy vs Time (NVT)')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.legend()
```

```python
#plot the data
plt.plot(npt[:,1], npt[:,4], label='NPT')
plt.title('Potential Energy vs Time (NPT)')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (kJ/mol)')
plt.legend()
```
