```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import numpy as np
import pandas as pd

df = pd.read_csv('fes.dat', sep='\s+', skiprows=9, header=None)
df.columns = ['phi', 'psi', 'free ene', 'der_phi', 'der_psi']
df

phi=df['phi'].to_numpy()
psi=df['psi'].to_numpy()
fes=df['free ene'].to_numpy()

phi_new=phi[:33]
psi_new=psi[::33]
fes_new=fes.reshape(18,33)


plt.contourf(phi_new,psi_new,fes_new,20)
plt.xlabel('$\phi$')
plt.ylabel('$\psi$')
plt.title('Free Energy Surface DialaA 2MD')
plt.colorbar(label='Free Energy (kJ/mol)')
```
