```python
#load mdtraj and numpy
import mdtraj as md
import numpy as np
```

```python
#read the md trajectory of beta amyloid
traj= md.load('../Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='1hz3_T310.start.nowater.gro')
```

```python
#compute the end-to-end distance at each frame
end_to_end = md.compute_distances(traj, [[0, traj.n_atoms-1]])
print(end_to_end)
```

```python
#compute the radius of gyration at each frame
radius_of_gyration = md.compute_rg(traj)
print(radius_of_gyration)
```

```python
#plot the end-end distance on the left axis and radius of gyration on the right axis vs time 
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()
ax1.plot(traj.time, end_to_end, 'y-')
ax1.set_xlabel('Time (ps)')
ax1.set_ylabel('End-to-End Distance (nm)', color='y')
ax2 = ax1.twinx()
ax2.plot(traj.time, radius_of_gyration, 'g-')
ax2.set_ylabel('radius of gyration (nm)', color='g')
plt.title('End-to-End Distance & Radius of Gyration vs Time')
plt.show()
```
