```python
#import numpy and matlibplot
import numpy as np
import matplotlib.pyplot as plt

#open numpy file and select data columns and rows
data = np.loadtxt('../Data/1hz3_T310.run.25000000.energy.xvg', skiprows=29)

#select data column for time coverting from pecoseconds to nanoseconds
time = data[:,0]/1000

#select data column for temperature
temp = data[:,4]

#plot data with labels and title
plt.plot(time, temp)
plt.xlabel('Time (ns)')
plt.ylabel('Temperature (K)')
plt.title('Temperature vs Time')

#display plot
plt.show()
```
